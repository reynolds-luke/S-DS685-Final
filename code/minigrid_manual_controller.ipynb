```python

# USE THIS CODE BLOCK TO PLAY AS A HUMAN!!
ENV_ID          = "MiniGrid-DoorKey-5x5-v0"

import gym
import pygame
from gym_minigrid.wrappers import RGBImgObsWrapper, ImgObsWrapper

ENV_ID = "MiniGrid-MultiRoom-N6-v0"
FPS = 10

KEY_MAP = {
    pygame.K_LEFT: 0,   # turn left
    pygame.K_RIGHT: 1,  # turn right
    pygame.K_UP: 2,     # move forward
    pygame.K_SPACE: 3,  # pick up
    pygame.K_d: 4,      # drop
    pygame.K_RETURN: 5, # toggle
    pygame.K_q: 6       # done
}

def play():
    env = gym.make(ENV_ID, render_mode="rgb_array")
    env = RGBImgObsWrapper(env)
    env = ImgObsWrapper(env)

    pygame.init()
    clock = pygame.time.Clock()

    obs, _ = env.reset()
    img = env.render()
    height, width = img.shape[:2]

    # ‚úÖ Show at exact render resolution (no zooming)
    screen = pygame.display.set_mode((width, height))
    pygame.display.set_caption(f"MiniGrid: {ENV_ID}")

    running = True
    while running:
        action = None
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.KEYDOWN:
                if event.key in KEY_MAP:
                    action = KEY_MAP[event.key]
                elif event.key == pygame.K_r:
                    obs, _ = env.reset()

        if action is not None:
            obs, reward, terminated, truncated, _ = env.step(action)
            print(reward)
            if reward > 0:
                print(f"‚úÖ Reward: {reward}")
            if terminated or truncated:
                print("üèÅ Episode finished. Press R to reset.")
                obs, _ = env.reset()

        img = env.render()  # already a full RGB array
        surf = pygame.surfarray.make_surface(img.transpose(1, 0, 2))  # [W, H, C]
        screen.blit(surf, (0, 0))
        pygame.display.flip()
        clock.tick(FPS)

    env.close()
    pygame.quit()

if __name__ == "__main__":
    play()
```
